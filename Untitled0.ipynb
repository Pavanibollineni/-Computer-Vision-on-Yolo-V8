```python
import cv2
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

cap = cv2.VideoCapture("busymovs.mp4")

fps = int(cap.get(cv2.CAP_PROP_FPS))
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter("output.mp4",
                      cv2.VideoWriter_fourcc(*'mp4v'),
                      fps, (w, h))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.track(
        frame,
        persist=True,
        tracker="botsort.yaml"   # 🔥 KEY FIX
    )

    if results[0].boxes is not None and results[0].boxes.id is not None:
        boxes = results[0].boxes.xyxy.cpu().numpy()
        ids = results[0].boxes.id.cpu().numpy()

        for box, track_id in zip(boxes, ids):
            x1, y1, x2, y2 = map(int, box)

            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

            cv2.putText(frame, f'ID {int(track_id)}',
                        (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.7, (0, 0, 255), 2)

    out.write(frame)

cap.release()
out.release()

print("✅ Stable tracking video saved")

```

StrongSORT improves object tracking by combining motion prediction and appearance matching. It ensures that objects are tracked accurately across video frames, even in challenging situations like occlusion or missed detections.

**Object Detection**

First, YOLO detects objects (like people, cars) in each frame of a video.

**Prediction (Kalman Filter)**

The algorithm predicts where an object will move next based on its previous position.

**Matching Objects**

It compares objects in the current frame with the previous frame using:

Position (movement)

Appearance (looks like clothes, color)

**Assigning IDs**

Each object is given a unique ID and tracked across frames.

**Handling Missing Objects**

If an object disappears for a few frames (like behind another object), StrongSORT can still track it when it reappears.

**⭐ Key Features**

Keeps same ID for the same object

Works well even when objects overlap or disappear

Uses both motion and appearance for better tracking

More accurate than older methods like SORT
